```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as wb
```

```python
x = ['PG','MSFT','GE','T']
df = pd.DataFrame()
for t in x:
    df[t] = wb.DataReader(t,data_source='yahoo',start='1-1-1995')['Adj Close']
```

```python
df.head()
```

```python
df.iloc[0]
```

```python
((df/df.iloc[0])*100).plot(figsize = (15,6))
plt.show()
```

Rate OF Return Of PortFolio

```python
returns = (df/df.shift(1))-1
```

```python
returns.head()
```

```python
weights = np.array([0.25,0.25,0.25,0.25])
```

```python
annual_return = returns.mean()*250
```

```python
annual_return
```

```python
np.dot(annual_return,weights)
```

```python
print (str(round(np.dot(annual_return,weights),5)*100),'%')
```
