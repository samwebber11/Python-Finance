```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as wb
```

```python
x = ['PG','^GSPC']
df = pd.DataFrame()
for t in x:
    df[t] = wb.DataReader(t,data_source = 'yahoo',start = '2013-1-1',end = '2018-1-1')['Adj Close']
```

```python
df.head()
```

```python
log_return = np.log(df/df.shift(1))
```

```python
log_return
```

```python
sec_return = log_return.cov()*250
sec_return
```

```python
market_price = log_return['^GSPC'].var()*250
```

```python
market_price
```

```python
cov_market = sec_return.iloc[0,1]
```

```python
cov_market
```

```python
beta = cov_market/market_price
```

```python
beta
```

```python
expected_return = 0.025 + beta*0.05
```

```python
expected_return
```
