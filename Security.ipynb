```python
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
x = ['PG','BEI.DE']
df = pd.DataFrame()
for t in x:
    df[t] = wb.DataReader(t,data_source='yahoo',start='1-1-2008')['Adj Close']
df
```

```python
df.head()
```

```python
security = np.log(df/df.shift(1))
security
```

```python
security['PG'].mean()*250
```

```python
security['PG'].std()*250 ** 0.5
```

```python
security[['PG','BEI.DE']].std()*250 ** 0.5
```

```python
security['PG'].var()*250
```

```python
security['BEI.DE'].var()*250
```

```python
cov_matrix = security.cov()*250
cov_matrix
```

```python
cor_matrix = security.corr()
cor_matrix
```

#This shows how two returns are related to each other not how prices are related to each other


# portfolio risk

portfolio Variance

```python
weights = np.array([0.5,0.5])
```

```python
port_var = np.dot(weights.T,np.dot(security.cov()*250,weights))
```

```python
port_var
```

```python
print (str(round(port_var,5)*100),'%')
```

diversifiable risk

```python
pg_var = security['PG'].var()*250
```

```python
bie_var = security['BEI.DE'].var()*250
```

```python
risk = port_var - (weights[0] **2*pg_var + weights[1]**2*bie_var)
```

```python
risk
```

```python
print (str(round(risk,5)*100),'%')
```

Non Diversifiable Risk

```python
nf1 = port_var - risk
nf1
```

```python
nf2 = weights[0] **2*pg_var + weights[1]**2*bie_var
```

```python
nf2
```

```python
nf1 == nf2
```
